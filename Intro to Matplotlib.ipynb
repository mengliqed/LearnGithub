Matplotlib has a simple MATLAB-style APIA(Application Programming Interface)

**The Object-Oriented API**

```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
# fig is a Figure instance, like a blank canvas
# ax is an AxesSubplot instance, like a plot to plot in

x = np.linspace(0, 10, 200)
y = np.sin(x)

ax.plot(x, y, 'b-', linewidth=2)
plt.show()
```

Now let's tweak the code and add more features.

```python
import numpy as up
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
x = np.linspace(0, 10, 200)
y = np.sin(x)

ax.plot(x, y, 'r-', linewidth=2, label=r'$y=\sin(x)', alpha=0.6)
ax.legend(loc='upper center')
ax.set_yticks([-1, 0, 1])
ax.set_title('Test plot')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from random import uniform

fig, ax = plt.subplots()
x = np.linspace(-4, 4, 150)
for i in range(3):
    m, s = uniform(-1, 1), uniform(1, 2)
    y = norm.pdf(x, loc=m, scale=s)
    current_label = r'$\mu = {0:.2f}$'.format(m)
    ax.plot(x, y, linewidth=2, alpha=0.6, label=current_label)
ax.legend()
plt.show()
```

**Multiple Subplots** to have multiple subplots in one figure.

```python
import matplotlib.pyplot as plt
from scipy.stats import norm
from random import uniform

num_rows, num_cols = 3, 2
fig, axes = plt.subplots(num_rows, num_cols, figsize=(8, 12))

for i in range(num_rows):
    for j in range(num_cols):
        m, s = uniform(-1, 1), uniform(1, 2)
        x = norm.rvs(loc=m, scale=s, size=100)
        axes[i, j].hist(x, alpha=0.6, bins=20)
        t = r'$\mu = {0:.1f}, \quad \sigma = {1:.1f}$'.format(m,s)
        axes[i, j].set_title(t)
        axes[i, j].set_xticks([-4, 0, 4])
        axes[i, j].set_yticks([])
plt.show()
```
